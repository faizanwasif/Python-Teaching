```python
import turtle
import math

# Set up the screen
window = turtle.Screen()
window.bgcolor("black")

# Create a turtle object
pen = turtle.Turtle()
pen.speed(3)  # Set a slower drawing speed

# Function to draw a grid
def draw_grid():
    pen.penup()
    pen.goto(-600, 0)
    pen.pendown()
    pen.color("white")

    for x in range(-600, 601, 100):
        pen.penup()
        pen.goto(x, -200)
        pen.pendown()
        pen.goto(x, 200)

    pen.penup()
    pen.goto(-600, 0)
    pen.pendown()

    for y in range(-200, 201, 100):
        pen.penup()
        pen.goto(-600, y)
        pen.pendown()
        pen.goto(600, y)

# Function to draw a sine wave
def draw_sine_wave(amplitude, frequency, wavelength):
    x_start = -600  # Extend the drawing range
    x_end = 600
    period = x_end - x_start

    pen.penup()
    pen.goto(x_start, amplitude * math.sin(2 * math.pi * frequency * (-period / 2)))
    pen.pendown()
    pen.color("yellow")
    pen.pensize(3)  # Increase the line width

    for x in range(x_start, x_end + 1, 1):  # Smoother curve
        y = amplitude * math.sin(2 * math.pi * frequency * (x - x_start) / period)
        pen.goto(x, y)

# Draw a grid
draw_grid()

# Draw a sine wave
draw_sine_wave(100, 1, 1200)  # Increased wavelength

# Close the drawing window on click
window.exitonclick()
```
